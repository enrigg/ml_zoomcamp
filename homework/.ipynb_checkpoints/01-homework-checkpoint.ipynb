```python
# Q7
innjoo = df[(df.Brand == "Innjoo")]
innjoo_selected = innjoo[["RAM", "Storage", "Screen"]]
innjoo_selected
```

```python
X = np.array(innjoo_selected)
X
```

```python
XTX = X.T.dot(X)
```

```python
XTX
```

```python
np.linalg.det(XTX)
```

```python
inver = np.linalg.inv(XTX)
inver
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100])
```

```python
inver_T = inver.dot(X.T)
inver_T
```

```python
w = inver_T.dot(y)
w
```
